## Cleaning, Standardizing, and Normalizing

```python
import pandas as pd
```

```python
nasa = pd.read_csv('neo.csv')
nasa.head()
```

```python
# drop unnecessary columns
dropcolumns = ['id', 'name', 'orbiting_body', 'sentry_object']
nasa = nasa.drop(columns=dropcolumns)
```

```python
# convert the index into a column to be used as a primary key

nasa.reset_index(inplace=True)
```

```python
# use get.dummies to convert T/F values into integers
nasa = pd.get_dummies(nasa, columns=['hazardous'])

#delete the redundant negative value column
nasa = nasa.drop(['hazardous_False'], axis=1)

#rename "hazardous_True" to "hazardous"
nasa.rename(columns={'hazardous_True': 'hazardous'}, inplace=True)

nasa.head()
```

## Exporting as CSV

```python
#export as csv
nasa.to_csv('neoclean.csv', index=False)
```
